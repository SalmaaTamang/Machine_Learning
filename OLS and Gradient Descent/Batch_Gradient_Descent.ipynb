```python
class Regressor:
    def __init__(self,learning_rate,epochs):
        
        self.coef = None
        self.intercept = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,x_train,y_train):
        #here at first initalize the coefficient and intercept
        self.intercept = 0
        #since when we deal will multiple linear regression we have numbers of coff value
        #so our task is to find how much coff is there. This can be done by finding no of cols in
        #train set. So we will make empty array of the train shape
        self.coef = np.ones(x_train.shape[1])
        
        
        for i in range(self.epochs):
                #we will update the val of intercept and coff
            y_hat = self.intercept + np.dot(x_train, self.coef)
            intercept_derivative = -2 * np.mean(y_train-y_hat)
            self.intercept = self.intercept - (self.lr*intercept_derivative)
                
            coef_derivative = -2*np.dot((y_train-y_hat),x_train)/x_train.shape[0]
            self.coef = self.coef -(self.lr*coef_derivative)
                
        print(self.intercept,self.coef)
    
    def predict(self, x_test):
                
        return np.dot(x_test,self.coef)+self.intercept
        
```

```python
gd = Regressor(0.8,1000)
```

```python
gd.fit(x_train,y_train)
```

```python
y_predict = gd.predict(x_test)
r2_score(y_test,y_predict)
```
